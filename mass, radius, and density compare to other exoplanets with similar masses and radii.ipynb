```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import astropy.io.ascii 
import astropy.constants as ac
import astropy.units as u
```

```python
style_revisions = {
 'axes.linewidth':  1.5, 
 'xtick.top' : True,         
 'ytick.right' :  True,
 'xtick.direction' : 'in',    
 'ytick.direction' : 'in', 
 'xtick.major.size' : 8,     
 'ytick.major.size' : 11,
 'xtick.minor.size' : 5.5,    
 'ytick.minor.size' : 5.5,      
 'font.size' : 12,              
 'figure.figsize' : [7.5, 4.5], 
 'lines.linewidth' : 2.      
}

# Update the matplotlib dictionary.
plt.rcParams.update(style_revisions)
```

```python
exo = astropy.io.ascii.read("PS_2023.02.19_10.15.49.csv")
exos = astropy.io.ascii.read("PS_2023.02.20_12.37.15.csv")
```

```python
Osirism = 0.71
Osirisr = 1.4
Osirisd = 0.31094
```

```python
Jupiterm = 1
Jupiterr = 1
Jupiterd = 1.326
```

```python
plt.plot(exos["pl_massj"], exos["pl_radj"], "b.", label='Exoplanets')
plt.plot(Jupiterm, Jupiterr, "o",color = "orange", marker='d', label='Jupiter')
plt.plot(Osirism, Osirisr, "r.", marker='s', label='Osiris')
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Mass (Jupiter Mass)")
plt.ylabel("Radius (Jupiter Radius)")
plt.legend(loc='best', prop={'size': 12})
```

```python
plt.plot(exos["pl_massj"], exos["pl_dens"], "b.", label='Exoplanets')
plt.plot(Osirism, Osirisd, "r.", marker='s', label='Osiris')
plt.plot(Jupiterm, Jupiterd,"o", color="orange", marker='d', label='Jupiter')
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Mass (Jupiter Mass)")
plt.ylabel("Density (g/cm^3)")
plt.legend(loc='best', prop={'size': 12})
```
